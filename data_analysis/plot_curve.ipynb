```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sa_modifier(x):
    mu = 2.230044
    sigma = 0.6526308
    x = np.maximum(x, mu)
    return np.exp(-0.5 * np.power((x - mu) / sigma, 2.)) 

def sc_modifier(x):
    mu = 2.698693
    sigma = 0.091844318
    x = np.maximum(x, mu)
    return np.exp(-0.5 * np.power((x - mu) / sigma, 2.)) 
```

```python
line_width = 5
font_size = 16

fig, ((ax1, ax2)) = plt.subplots(1, 2, figsize=(13, 5))

x = np.arange(1, 6, 0.01)
y = sa_modifier(x)

ax1.plot(x, y, linewidth=line_width)
ax1.grid()
ax1.set_ylabel('Output value of multiplier', size=font_size)
ax1.set_xlabel('Input value of SA_Score', size=font_size)

x = np.arange(1, 5, 0.01)
y = sc_modifier(x)

ax2.plot(x, y, linewidth=line_width)
ax2.grid()
ax2.set_xlabel('Input value of SCScore', size=font_size)


plt.savefig('./figure/multipliers.png')
```
